```python
from propose.datasets.human36m.Human36mDataset import Human36mDataset
from propose.poses.human36m import Human36mPose

import matplotlib.pyplot as plt
import numpy as np
```

```python
train_dataset = Human36mDataset(
    dirname="/data/human36m/processed",
    mpii=True,
    use_variance=True,
    occlusion_fractions=[],
    num_samples=10000
)
```

```python
test_dataset = Human36mDataset(
    dirname="/data/human36m/processed",
    mpii=True,
    use_variance=True,
    occlusion_fractions=[],
    test=True,
)
```

```python
hard_dataset = Human36mDataset(
    dirname="/data/human36m/processed",
    mpii=True,
    use_variance=True,
    occlusion_fractions=[],
    hardsubset=True,
)
```

```python
occlusions = train_dataset.occlusions.astype(int).sum(0)

marker_names = Human36mPose(np.zeros((16, 2))).marker_names[1:]
del marker_names[8]

plt.style.use('default')
plt.bar(marker_names, occlusions)
plt.xticks(rotation=90)
plt.show()
```

```python
occlusions = test_dataset.occlusions.astype(int).sum(0)

marker_names = Human36mPose(np.zeros((16, 2))).marker_names[1:]
del marker_names[8]

plt.style.use('default')
plt.bar(marker_names, occlusions)
plt.xticks(rotation=90)
plt.show()
```
